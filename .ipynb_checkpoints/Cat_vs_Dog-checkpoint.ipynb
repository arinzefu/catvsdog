```python
import cv2
import imghdr
import torch
import torch.nn as nn
import torch.optim as optim
import os 
from torch.utils.data import random_split, DataLoader
from torchvision import transforms
import torch.nn.functional as F
import numpy as np
from matplotlib import pyplot as plt
import torchvision.datasets as datasets
```

```python
Train_set = 'train'
Test_set = 'test'
print(os.listdir(Train_set))
print(os.listdir(Test_set))
```

```python
total_images = 0  # initialize counter variable
for image_class in os.listdir(Train_set):
    for image in os.listdir(os.path.join(Train_set, image_class)):
        total_images += 1  # increment counter for each file
print("Total number of Train images:", total_images)
```

```python
total_images = 0  # initialize counter variable
for image_class in os.listdir(Test_set):
    for image in os.listdir(os.path.join(Test_set, image_class)):
        total_images += 1  # increment counter for each file
print("Total number of Test images:", total_images)
```

```python
transform = transforms.Compose([
    transforms.RandomResizedCrop(224),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])
```

```python
train = datasets.ImageFolder(Train_set, transform=transform)
test =  datasets.ImageFolder(Test_set, transform=transform)
```

```python

batch_size = 16
train_loader = DataLoader(train, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test, batch_size=batch_size, shuffle=False)
```

```python
# Get the first batch
first_batch = next(iter(train_loader))

# Print the first batch
print(first_batch)
```

```python
# Define the device (GPU or CPU)
device = ('cuda')
```

```python

import torchvision.models as models

num_classes = 2

class ImageClassificationModule(nn.Module):
    def __init__(self, dropout_rate=0.1):
        super(ImageClassificationModule, self).__init__()
        self.resnet = models.resnet18(pretrained=True)
        num_ftrs = self.resnet.fc.in_features
        self.resnet.fc = nn.Sequential(
            nn.Linear(num_ftrs, num_classes),
            nn.Dropout(p=dropout_rate)
        )

    def forward(self, x):
        x = self.resnet(x)
        return x
```

```python
# Initialize the model and move it to the device
from torchsummary import summary
model = ImageClassificationModule().to(device)

summary(model, input_size=(3, 224, 224))
```

```python
import torch.optim as optim
```

```python
# Define the loss function and optimizer
criterion = nn.BCEWithLogitsLoss()
optimizer = optim.SGD(model.parameters(), lr=0.001, momentum=0.9)
```

```python
# Train the model
for epoch in range(3):
    train_loss = 0.0
    train_correct = 0
    val_loss = 0.0
    val_correct = 0

    # Training loop
    model.train()
    for inputs, labels in train_loader:\
        inputs, labels = inputs.to(device), labels.to(device)
    optimizer.zero_grad()
    outputs = model(inputs)


    loss = criterion(outputs, labels)
    loss.backward()
    optimizer.step()

    train_loss += loss.item()
    _, predicted = torch.max(outputs.data, 1)
    train_correct += (predicted == labels).sum().item()
```
